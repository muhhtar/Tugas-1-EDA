```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df = pd.read_excel('Maths.csv')
df.head()
```

```python
print("Ukuran Dataset:", df.shape)
print("\nKolom:")
print(df.columns)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
df_numeric = df.select_dtypes(include=['number'])
df_numeric.head()
```

```python
df_numeric.describe()
```

```python
df_numeric.hist(figsize=(12,8))
plt.suptitle('Distribusi Data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df_numeric)
plt.title('Deteksi Outlier')
plt.show()
```

```python
target = df_numeric.columns[-1]
print("Target:", target)
```

```python
for col in df_numeric.columns:
    if col != target:
        plt.figure(figsize=(5,3))
        sns.scatterplot(x=df_numeric[col], y=df_numeric[target])
        plt.title(f'{col} vs {target}')
        plt.show()
```
